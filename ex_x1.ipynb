```python
df = pd.read_csv('https://github.com/iam-fern/Research/raw/main/gen_SAS.csv', index_col=['Replicate', 'no'])
df
```

```python
# ลบ  colum
df.drop(['Unnamed: 13'], axis=1, inplace=True)
df
```

```python
# ลบข้อมูลที่มี missing values
table = df.dropna()
table
```

```python
table.reset_index(inplace=True)
table
```

```python
table.set_index(['Replicate', 'no'], inplace=True)
```

```python
table
```

```python
# แสดงค่าของ Replicate ที่ 1.0
new_table = table.loc[5.0]
new_table
```

```python
# เรียง data จากน้อยไปมากตาม x1
new_table_sort = new_table.sort_values('x1')
new_table_sort
```

```python
new_table_sort.reset_index(inplace=True)
new_table_sort
```

```python
new_table_sort.drop(['no', 'id', 'j', 'a', 'b', 'c', 'z', 'x2', 'x3', 'E'], axis=1, inplace=True)
new_table_sort
```

```python
table1 = new_table_sort.iloc[:int(0.4*new_table_sort.shape[0]) , :]
table1
```

```python
table2 = new_table_sort.iloc[int((1-0.4)*new_table_sort.shape[0]): , :]
table2
```

```python
# define respones variable
y = table1['y']

# define predictor variable
x = table1[['x1']]

# add constant to predictor variable
x = sm.add_constant(x)

# fit linear regression model
model1 = sm.OLS(y, x).fit()
```

```python
DF1 = model1.df_resid
DF1
```

```python
SSE1 = model1.ssr
SSE1
```

```python
model1.mse_resid
```

```python
# define respones variable
y = table2['y']

# define predictor variable
x = table2[['x1']]

# add constant to predictor variable
x = sm.add_constant(x)

# fit linear regression model
model2 = sm.OLS(y, x).fit()
```

```python
f = model2.mse_resid/model1.mse_resid
f
```
