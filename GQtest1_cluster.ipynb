```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
df = pd.read_csv('https://github.com/iam-fern/Research/raw/main/gen_SAS.csv', index_col=['Replicate', 'no'])
df
```

```python
# ลบ  colum
df.drop(['Unnamed: 13'], axis=1, inplace=True)
df
```

```python
# ลบข้อมูลที่มี missing values
table = df.dropna()
table
```

```python
table.reset_index(inplace=True)
table
```

```python
table.set_index(['Replicate', 'no'], inplace=True)
```

```python
table
```

```python
# แสดงค่าของ Replicate ที่ 1.0
new_table = table.loc[2.0]
new_table
```

```python
# เรียง data จากน้อยไปมากตาม x1
#new_table_sort = new_table.sort_values('x1')
#new_table_sort
```

```python
# show index
new_table.reset_index(inplace=True)
new_table
```

```python
new_table.drop(['no', 'id', 'j', 'a', 'b', 'c', 'z', 'E'], axis=1, inplace=True)
new_table
```

```python
new_table.head()
```

```python
# แบ่งข้อมูลออกเป็น 2 กลุ่มด้วย k-mean

#Set number of clusters
x = new_table[["x1", "x2", "x3"]]
k_means = KMeans(n_clusters=2)

#Run the clustering algorithm
model = k_means.fit(x)
model

#Generate cluster predictions and store in y_kmeans
#km.fit(x)
y_kmeans = k_means.predict(x)
y_kmeans
```

```python
# kmean_clusters.astype(float)
kmean_clusters = pd.DataFrame({"clusters" : y_kmeans})
kmean_clusters
```

```python
# เชื่อมข้อมูลทั้ง 2 DataFrame เข้าด้วยกันด้วยคำสั่ง pd.merge
data_clusters = pd.merge(new_table, kmean_clusters, left_on='x1', right_on='clusters', left_index=True, right_index=True)
data_clusters
```

```python
data_sort_clusters = data_clusters.sort_values('clusters')
data_sort_clusters
```

```python
print(data_clusters.groupby('clusters').groups)
```

```python
data_sort_clusters.reset_index(inplace=True)
data_sort_clusters
```

```python
print(data_sort_clusters.groupby('clusters').groups)
```

```python
data_sort_clusters.columns
```

```python
import pandas as pd
import numpy as np
import time
import statsmodels.api as sm
from sklearn import linear_model
```

```python
data_cluster1 = data_sort_clusters[data_sort_clusters.clusters == 1]
data_cluster1

y = data_cluster1[['y']]
x = data_cluster1[['x1', 'x2', 'x3']]
x = sm.add_constant(x) # adding a constant
model1 = sm.OLS(y, x).fit()
```

```python
print_model1 = model1.summary()
print(print_model1)
```

```python
DF1 = model1.df_resid
DF1
```

```python
SSR1 = model1.ssr
SSR1
```

```python
MSE1 = model1.mse_resid
MSE1
```

```python
data_cluster0 = data_sort_clusters[data_sort_clusters.clusters == 0]
data_cluster0

y = data_cluster0[['y']]
x = data_cluster0[['x1', 'x2', 'x3']]
x = sm.add_constant(x)
model0 = sm.OLS(y, x).fit()
```

```python
#print_model0 = model0.summary()
#print(print_model0)
```

```python
DF0 = model0.df_resid
DF0
```

```python
SSR0 = model0.ssr
SSR0
```

```python
MSE0 = model0.mse_resid
MSE0
```

```python
GQ_test_clusters = MSE1 / MSE0
GQ_test_clusters
```

```python
# เปรียบเทียบค่าสถิติกับค่าวิกฤต
if GQ_test_clusters > 2.9130:
  print('Reject005=0 : Heteroscedasticity ')
else:
  print('Reject005=1 : Homoscedasticity ')
```
