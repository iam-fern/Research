```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.cluster import KMeans
```

```python
df = pd.read_csv('https://github.com/iam-fern/Research/raw/main/gen_SAS.csv', index_col=['Replicate', 'no'])
df
```

```python
# ลบ  colum
df.drop(['Unnamed: 13'], axis=1, inplace=True)
df
```

```python
# ลบข้อมูลที่มี missing values
table = df.dropna()
table
```

```python
table.reset_index(inplace=True)
table
```

```python
table.set_index(['Replicate', 'no'], inplace=True)
```

```python
table
```

```python
# แสดงค่าของ Replicate ที่ 1.0
new_table = table.loc[1.0]
new_table
```

```python
# เรียง data จากน้อยไปมากตาม x1
new_table_sort = new_table.sort_values('x1')
new_table_sort
```

```python
new_table_sort.reset_index(inplace=True)
new_table_sort
```

```python
new_table_sort.drop(['no', 'id', 'j', 'a', 'b', 'c', 'z', 'E'], axis=1, inplace=True)
new_table_sort
```

```python
new_table_sort.head()
```

```python
new_table_sort.y.unique()
```

```python
# แบ่งข้อมูลออกเป็น 2 กลุ่มด้วย k-mean
x = new_table_sort[["x1", "x2", "x3"]]
km = KMeans(n_clusters=2)
km.fit(x)
y_kmeans = km.predict(x)
y_kmeans
```
