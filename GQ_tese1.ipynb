```python
df = pd.read_csv('https://github.com/iam-fern/Research/raw/main/gen_SAS.csv', index_col=['Replicate', 'no'])
df
```

```python
# การเช็คว่าเจอ missing value หรือไม่ ถ้าไม่มีก็จะขึ้นว่า false คือไม่เจอ
df.isnull()
```

```python
# ลบ  colum
df.drop(['Unnamed: 13'], axis=1, inplace=True)
df
```

```python
# ลบข้อมูลที่มี missing values
table = df.dropna()
table
```

```python
table.reset_index(inplace=True)
table
```

```python
table.set_index(['Replicate', 'no'], inplace=True)
```

```python
table
```

```python
# แสดงค่าของ Replicate ที่ 1.0
new_table = table.loc[1.0]
new_table
```

```python
# เรียง data จากน้อยไปมากตาม x1
new_table_sort = new_table.sort_values('x1')
new_table_sort
```

```python
new_table_sort.reset_index(inplace=True)
new_table_sort
```

```python
new_table_sort.drop(['no', 'id', 'j', 'a', 'b', 'c', 'z', 'E'], axis=1, inplace=True)
new_table_sort
```

```python
# แบ่งข้อมูลกลุ่มที่ 1
table1 = new_table_sort.iloc[:13 , :]
table1
```

```python
import statsmodels.api as sm
```

```python
# define respones variable
y = table1['y']
```

```python
# define predictor variable
x = table1[['x1', 'x2', 'x3']]
```

```python
# add constant to predictor variable
x = sm.add_constant(x)
```

```python
# fit linear regression model
model = sm.OLS(y, x).fit()
```

```python
# view model summary
print(model.summary())
```

```python
# หาค่า df กลุ่ม 1
DF1 = print(model.df_resid)
DF1
```

```python
# หาค่า SSE1 : Residual sum of squares
SSE1 = print(model.ssr)
SSE1
```

```python
# แบ่งข้อมูลกลุ่มที่ 2
table2 = new_table_sort.iloc[17: , :]
table2
```

```python
y = table1['y']
x = table1[['x1', 'x2', 'x3']]
x = sm.add_constant(x)
model = sm.OLS(y, x).fit()
print(model.summary())
```

```python
# หาค่า df กลุ่ม 2
DF2 = print(model.df_resid)
DF2
```

```python
# หาค่า SSE2
SSE2 = print(model.ssr)
SSE2
```
