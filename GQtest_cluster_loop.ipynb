```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

import statistics as stats
import scipy.stats as ss
import scipy.stats as st
from scipy.stats import f_oneway
from scipy import stats
import scipy.stats as stats
import time
import statsmodels.api as sm

from statsmodels.formula.api import ols
```

```python
df = pd.read_csv('https://github.com/iam-fern/Research/raw/main/gen_SAS.csv', index_col=['Replicate', 'no'])
df
```

```python
# ลบ  colum
df.drop(['Unnamed: 13'], axis=1, inplace=True)
df
```

```python
# ลบข้อมูลที่มี missing values
table = df.dropna()
table
```

```python
table.reset_index(inplace=True)
table
```

```python
table.set_index(['Replicate', 'no'], inplace=True)
```

```python
table
```

```python
def get_result(run_replicate):
    new_table = table.loc[run_replicate]

    # เรียง data จากน้อยไปมากตาม x1
    #new_table_sort = new_table.sort_values('x1')
    #new_table_sort

    # show index
    new_table.reset_index(inplace=True)

    new_table.drop(['no', 'id', 'j', 'a', 'b', 'c', 'z', 'E'], axis=1, inplace=True)

    #new_table1 = new_table.astype(float)
    #new_table1

    # แบ่งข้อมูลออกเป็น 2 กลุ่มด้วย k-mean
    x = new_table[["x1", "x2", "x3"]]
    km = KMeans(n_clusters=2)
    km.fit(x)
    y_kmeans = km.predict(x)


    # kmean_clusters.astype(float)
    kmean_clusters = pd.DataFrame({"clusters" : y_kmeans})

    # เชื่อมข้อมูลทั้ง 2 DataFrame เข้าด้วยกันด้วยคำสั่ง pd.concat
    # data_clusters = pd.merge(new_table, kmean_clusters, left_on='x1', right_on='clusters', left_index=True, right_index=True)
    data_clusters = pd.concat([new_table, kmean_clusters], axis=1)

    data_sort_clusters = data_clusters.sort_values('clusters')

    # print(data_clusters.groupby('clusters').groups)

    data_sort_clusters.reset_index(inplace=True)

    # print(data_sort_clusters.groupby('clusters').groups)

    # data_sort_clusters.columns

    #for chue, klum in data_sort_clusters.groupby('clusters'):
        #print(chue)
        #print(klum)

    data_cluster1 = data_sort_clusters[data_sort_clusters.clusters == 1]

    y = data_cluster1[['y']]
    x = data_cluster1[['x1', 'x2', 'x3']]
    x = sm.add_constant(x)
    model1 = sm.OLS(y, x).fit()

    DF1 = model1.df_resid

    SSR1 = model1.ssr

    MSE1 = model1.mse_resid

    data_cluster0 = data_sort_clusters[data_sort_clusters.clusters == 0]

    y = data_cluster0[['y']]
    x = data_cluster0[['x1', 'x2', 'x3']]
    x = sm.add_constant(x)
    model0 = sm.OLS(y, x).fit()

    DF0 = model0.df_resid

    SSR0 = model0.ssr

    MSE0 = model0.mse_resid

    # คำนวณค่าสถิติ
    GQ_test_clusters = MSE1 / MSE0

    alpha = 0.05

    df1 = DF1
    df0 = DF0

    # คำนวณค่า p-value
    p_value = 2.0*(ss.f.cdf(GQ_test_clusters, df1, df0)


# เปรียบเทียบค่า p-value กับค่า alpha
    if p_value < alpha:
        result_test='Reject005=0 : Heteroscedasticity '
        result_test01=1
    else:
        result_test='Reject005=1 : Homoscedasticity '
        result_test01=0
    return df0, df1, SSR0, SSR1, MSE0, MSE1, GQ_test_clusters, result_test, p_value, result_test01

# สร้าง list ของ replicates
list_replicate=list({i[0] for i in table.index})
list_replicate.sort()

# สร้าง list เพื่อเก็บค่า ทุกรอบที่ run
list_df0=[]
list_df1=[]
list_SSR0=[]
list_SSR1=[]
list_MSE0=[]
list_MSE1=[]
list_GQ_test_clusters=[]
list_result_test=[]
list_p_value=[]
list_result_test01=[]

# ใช้ for loop สำหรับการ run ทุก replicates
for run_replicate in list_replicate:
    df0, df1,SSR0, SSR1, MSE0, MSE1, GQ_test_clusters, result_test, p_value, result_test01=get_result(run_replicate)

    list_df0.append(df0)
    list_df1.append(df1)
    list_SSR0.append(SSR0)
    list_SSR1.append(SSR1)
    list_MSE0.append(MSE0)
    list_MSE1.append(MSE1)
    list_GQ_test_clusters.append(GQ_test_clusters)
    list_result_test.append(result_test)
    list_p_value.append(p_value)
    list_result_test01.append(result_test01)
```

```python
# สร้างตารางผลลัพธ์
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)
result=pd.DataFrame({"Replicate":list_replicate, "df0":list_df0, "df1":list_df1, "SSR0":list_SSR0,
                     "SSR1":list_SSR1, "MSE0":list_MSE0, "MSE1":list_MSE1,
                     "GQ_test_clusters":list_GQ_test_clusters, "p_value":list_p_value,
                     "Result_test":list_result_test})
result.set_index(['Replicate'], inplace=True)

result.astype(str)
```
